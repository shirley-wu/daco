```python
# Unzip databases-dict/

import json
import pickle
import pandas as pd

SPLIT = 'validation'
INDEX = 0

with open(SPLIT + '.json') as f:
    data = json.load(f)
data_item = data[INDEX]

for msg in data_item['messages']:
    print("-" * 10, msg['role'])
    print(msg['content'])
    print()

with open('databases-dict/{}.pkl'.format(data_item['table_id']), 'rb') as f:
    db = pickle.load(f)
print("Database title", db['title'])
for key, value in db['database'].items():
    print("Table", key)
    print(pd.DataFrame.from_dict(value))
    print()
```
